# Move mouse

```python
pg.moveTo(500, 500)
```

```python
# move mouse 10px down wrt current position. This is (x, y)
pg.move(0, 10)
```

```python
# duration is the time required to reach at xy position, tween I think it is animation.
pg.moveTo(0, 0, duration=3, tween=pg.easeInOutQuad)
```

```python
pg.moveRel(50, 70, duration=1)  # move cursor relative to current position i.e move 50px in x direction and 70px in y direction
#                                 assuming current position of cursor as zero
```

```python
# diffence between drag and move is, we use drag to select folders and move to just move the cursor
```

```python
pg.dragTo(20, 20, duration=3) # drag mouse to XY
```

```python
pg.dragRel(30, 30, duration=4) # drag mouse relative to its current position
```

# Tween
**https://pypi.org/project/PyTweening/**

```python
pg.easeInQuad #functions to have the mouse cursor start off moving slowly and then speeding up towards the destination.

pg.easeOutQuad #reverse: the mouse cursor starts moving fast but slows down as it approaches the destination.

pg.easeOutElastic #will overshoot the destination and “rubber band” back and forth until it settles at the destination.
```
